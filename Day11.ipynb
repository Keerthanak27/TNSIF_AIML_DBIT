```python
import os
os.environ["OMP_NUM_THREADS"] = "1"   # try to avoid MKL/KMeans warning on Windows

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
```

```python
#Load CSV
csv_path = r"C:\Users\keert\OneDrive\Desktop\dbit\Mufg-AIML\student_kmeans_dataset.csv"
df = pd.read_csv(csv_path)
print("Dataset shape:", df.shape)
display(df.head()) 
```

```python
# KMeans and cluster column
k = 2
X = df[["Study_Hours", "Exam_Score"]]

kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df[f"Cluster_K{k}"] = kmeans.fit_predict(X)

print(f"KMeans completed for K={k}. Cluster counts:")
print(df[f"Cluster_K{k}"].value_counts().sort_index())

# Plot DataFrame
plt.figure(figsize=(8,6))
plt.scatter(df["Study_Hours"], df["Exam_Score"],
            c=df[f"Cluster_K{k}"], cmap="viridis", s=100, edgecolor="k", alpha=0.9)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=220, label="Centroids")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title(f"K-Means Clustering (K={k})")
plt.legend()
plt.grid(True)
plt.show()

display(df)
```
